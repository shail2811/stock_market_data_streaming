```
from pyspark.sql.functions import from_json, col, avg
from pyspark.sql.types import StructType, StructField, StringType, DoubleType
from pyspark.sql.functions import pandas_udf
from pyspark.sql.types import DoubleType
from pyspark.sql.functions import *

streaming_schema = StructType([
    StructField('Ticker', StringType(), True),
    StructField('Close', StringType(), True),
    StructField('sma_10', StringType(), True),
    StructField('ema_20', StringType(), True),
    StructField('rsi_14', StringType(), True),
    StructField('Signal', StringType(), True)
])


streaming_df = (
    spark
    .readStream
    .format("kafka")
    .option("kafka.bootstrap.servers", "54.245.47.165:9092")
    .option("subscribe", "demo_testing1")
    .load()
    .select(from_json(col("value").cast("string"), streaming_schema).alias("parsed_value"))
    .select("parsed_value.*")
)
```

```
# Start the streaming query and write the output to a sink 
query = (streaming_df 
  .writeStream 
  .outputMode("append") 
  .format("console") 
  .start())

display(streaming_df)
```
